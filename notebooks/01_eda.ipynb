# 01 — Exploratory Data Analysis on OULAD
Quick look at the seven OULAD tables, distributions, missingness, and class balance.


```python
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent))
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
DATA = Path('../data/raw')
student = pd.read_csv(DATA/'studentInfo.csv')
vle = pd.read_csv(DATA/'studentVle.csv')
print('students:', student.shape, 'vle:', vle.shape)
student.head()
```

```python
student['final_result'].value_counts(normalize=True).plot(kind='bar', title='Outcome distribution')
plt.tight_layout()
```

```python
click_by_day = vle.groupby('date')['sum_click'].sum()
click_by_day.plot(figsize=(10,3), title='Total VLE clicks per relative day')
```
